## Contexte
L'entreprise de produits pharmaceutiques HumanForYou basée en Inde emploie environ 4000 personnes. Cependant, chaque année elle subit un turn-over d'environ 15% de ses employés nécessitant de retrouver des profils similaires sur le marché de l'emploi.
La direction trouve que ce niveau de turn-over n'est pas bon pour l'entreprise car :
- Les projets sur lesquels étaient les employés quittant la société prennent du retard ce qui nuit à la réputation de l'entreprise auprès de ses clients et partenaires.
- Un service de ressources humaines de taille conséquente doit être conservé car il faut avoir les moyens de trouver les nouvelles recrues.
- Du temps est perdu à l'arrivée des nouveaux employés car ils doivent très souvent être formés et ont besoin de temps pour devenir pleinement opérationnels dans leur nouvel environnement.

## Importation des données

Nous avons reçu un ensemble de données du service des ressources humaines sous forme de fichiers CSV. Ces données, entièrement anonymisées, garantissent l’identification unique de chaque employé grâce à un identifiant **EmployeeID**, qui est constant à travers tous les fichiers.

Nous allons commencer par importer ces fichiers afin de les préparer pour l’analyse.

```python
from urllib.request import urlretrieve
import pandas as pd
import os

DATASET_URL = "https://raw.githubusercontent.com/adilelbaraka/IAproject/main/dataset/"

DATASET_PATH = "/content/datasets/"
if not os.path.exists(DATASET_PATH):
  os.mkdir(DATASET_PATH)


# Import des fichiers csv (datasets) stockés dans un repo github
def import_data(file_name):
  urlretrieve(DATASET_URL+file_name, DATASET_PATH+file_name)
  return pd.read_csv(DATASET_PATH+file_name)

survey = import_data("employee_survey_data.csv")
general = import_data("general_data.csv")
in_time = import_data("in_time.csv")
out_time = import_data("out_time.csv")
manager_survey = import_data("manager_survey_data.csv")
```

## Traitement des données des fichiers "in_time" et "out_time"

Les fichiers `in_time.csv` et `out_time.csv` ne sont pas directement exploitables, car notre analyse ne porte pas sur les horaires d'arrivée et de départ des employés.

Cependant, après examen de ces données, nous avons identifié **deux indicateurs clés** à extraire :

- **La durée moyenne de travail journalière** pour l'année **2015**.
- **Le nombre total d'absences** sur la même période.

Ces informations seront consolidées dans un **nouveau fichier**, `duration_data.csv`, qui regroupera ces données essentielles pour l'analyse.

```python
def calculate_duration(in_time, out_time):
    for col in in_time.columns[1:]:
        in_time[col] = pd.to_datetime(in_time[col])
        out_time[col] = pd.to_datetime(out_time[col])

    # Calculer la durée
    duration = in_time.copy()
    for col in in_time.columns[1:]:
        duration[col] = out_time[col] - in_time[col]
        duration[col] = duration[col].dt.total_seconds() / 3600  # Convertir en heures

    return duration


# Calcul de la durée
duration_data = calculate_duration(in_time, out_time)

# Calculer séparément les nouvelles colonnes
average_duration = duration_data.iloc[:, 1:].mean(axis=1, skipna=True)
absences_count = duration_data.iloc[:, 1:-1].isna().sum(axis=1)

# Concaténer les nouvelles colonnes avec le DataFrame original
duration_data = pd.concat([duration_data, average_duration.rename('Average Duration'),
                           absences_count.rename('Absences Count')], axis=1)

duration_data = duration_data.rename(columns={'Unnamed: 0': "EmployeeID"})

duration_data = duration_data[['EmployeeID','Average Duration', 'Absences Count']]

# Exportation des données
duration_data.to_csv("duration_data.csv", index=False)
```

# 🔍 Identification et élimination des variables non explicatives

#### 📌 Objectif : Comprendre la non-explicativité
Dans le cadre de notre projet, il est essentiel de déterminer quels attributs influencent significativement l'attrition, qui est le paramètre central de notre analyse. L’objectif est donc d’identifier et d’éliminer les variables non explicatives, c'est-à-dire celles qui n’apportent pas de valeur prédictive pour l’attrition.

#### 🧩 Critère de non-explicativité
Une variable est considérée comme **non explicative** si elle présente une distribution similaire entre les deux groupes d’attrition ("Oui" et "Non"). Une telle variable n’a pas de variation significative entre ces groupes et n’est donc pas un bon indicateur pour prédire l’attrition.

#### 📊 Analyse des variables numériques
Pour identifier les variables numériques non explicatives, nous utiliserons des **boîtes à moustaches** (boxplots). Ces visualisations permettent de comparer les distributions des variables entre les employés qui quittent l’entreprise et ceux qui restent.

- Si les distributions d’une variable sont similaires entre les deux groupes, elle sera considérée comme **non significative** pour l’analyse de l’attrition et pourra être éliminée.
- En revanche, si une variable montre des différences marquées entre les groupes, elle sera conservée comme attribut explicatif.

Cette approche garantit que seules les variables ayant un impact significatif sur l’attrition seront prises en compte dans les modèles prédictifs.

### 🔍 Variables catégorielles non explicatives

Pour les variables catégorielles, nous appliquerons une méthodologie différente. Seules les variables montrant des différences significatives dans leur distribution entre les employés ayant quitté l’entreprise et ceux qui sont restés seront retenues. Ces différences reflètent leur potentiel explicatif en lien avec l’attrition.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Liste des attributs catégoriels
cat_attribs = ["BusinessTravel", "Department", "Gender", "JobRole", "MaritalStatus", "Over18", "EducationField"]

# Configuration pour afficher plusieurs graphiques
fig, axes = plt.subplots(nrows=len(cat_attribs), ncols=1, figsize=(10, 6*len(cat_attribs)))
fig.tight_layout(pad=6.0)

for i, attribut in enumerate(cat_attribs):
    # Calculer les pourcentages
    percentage_df = complete_csv.groupby(['Attrition', attribut]).size().unstack(fill_value=0).apply(lambda x: x / x.sum() * 100, axis=1)

    # Tracer le diagramme en barres sur le subplot correspondant
    ax = axes[i]
    percentage_df.plot(kind='bar', stacked=True, ax=ax)
    ax.set_title(f'Pourcentage de {attribut} par Attrition')
    ax.set_ylabel('Pourcentage (%)')
    ax.set_xlabel('Attrition')

    # Afficher les pourcentages sur les barres
    for n, x in enumerate([*percentage_df.index.values]):
        for (proportion, y_loc) in zip(percentage_df.loc[x], percentage_df.loc[x].cumsum()):
            ax.text(x=n-0.1, y=y_loc - (proportion / 2), s=f'{proportion:.1f}%', color='white', fontsize=10)

    ax.set_xticklabels(['No', 'Yes'])
    ax.legend(title=attribut, bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

## Corrélation entre les attributs numériques

L'identification des corrélations entre variables permet de mieux comprendre leurs relations. Une **corrélation positive** élevée indique qu'une augmentation d'une variable est souvent associée à une augmentation de l'autre, tandis qu'une **corrélation négative** élevée signifie qu'une augmentation de l'une correspond généralement à une diminution de l'autre. Une corrélation proche de 0 indique qu'il n'y a pas de relation linéaire.

Pour visualiser ces corrélations, une **matrice de corrélation** est utilisée, souvent représentée sous forme de **heatmap** avec Seaborn. Les coefficients de corrélation varient entre :
- **+1** : Forte corrélation positive.
- **-1** : Forte corrélation négative.
- **0** : Absence de corrélation.

Dans la heatmap, les relations sont colorées pour faciliter l’interprétation :
- Les **couleurs chaudes** représentent des corrélations positives.
- Les **couleurs froides** signalent des corrélations négatives.

Cette analyse est particulièrement utile pour :
- Repérer les variables fortement liées entre elles.
- Identifier les problèmes de multicolinéarité dans les modèles de régression.
- Aider à sélectionner les variables pertinentes pour la modélisation.

### Transformation des données catégorielles non ordinales pour l'analyse

Les données catégorielles regroupent des éléments en différents groupes distincts. Dans notre ensemble de données, des colonnes telles que **`Department`** et **`JobRole`** illustrent ce type de données. Elles sont qualifiées de **non ordinales** lorsqu'il n'existe aucun ordre ou hiérarchie naturelle entre les catégories. Par exemple, pour la colonne **`Department`**, il n'y a aucune logique à considérer "Sales" comme supérieur ou inférieur à "Research & Development".

### Transformation via One-Hot Encoding

Pour rendre ces données utilisables dans les modèles de machine learning, qui nécessitent des données numériques, nous appliquons la méthode du **One-Hot Encoding**. Cette technique convertit les catégories en vecteurs numériques composés de **0** et de **1**, tout en évitant d'introduire une hiérarchie artificielle.

Voici comment cela fonctionne :
- Chaque catégorie unique d'une colonne est transformée en une nouvelle colonne distincte.
- Pour chaque observation, la colonne correspondant à sa catégorie sera marquée par un **1**, tandis que les autres colonnes seront remplies de **0**.
- Ce processus est appliqué à toutes les colonnes catégorielles concernées, assurant ainsi une représentation claire et exploitable des données non ordinales.

Cette transformation garantit que les modèles de machine learning traitent les données catégorielles sans biais lié à une relation d’ordre inexistante.

### Création de la pipeline

Nous pouvons formater les données de notre pipeline avec nos données formatées:

```python
# Importation des bibliothèques nécessaires
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np

data = complete_csv.drop("Attrition", axis=1)

# Sélection des caractéristiques numériques
donnees_num = data.select_dtypes(include=[np.number])
num_attribs = list(donnees_num)

# Les caractéristiques catégorielles ont déjà été sélectionnées

# Filtrage des attributs numériques, en enlevant ceux spécifiés dans to_remove_num
num_attribs = [attr for attr in num_attribs if attr not in to_remove_num]
print(num_attribs)

# Pipeline pour les données numériques
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),  # Remplissage des valeurs manquantes par la médiane
    ('std_scaler', StandardScaler()),               # Standardisation des caractéristiques
])

# Pipeline pour les données catégorielles
cat_pipeline = Pipeline([
    ('onehot_encoder', OneHotEncoder()),            # Encodage One-Hot pour les caractéristiques catégorielles
])

# Assemblage des pipelines numériques et catégorielles dans un seul transformateur
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),             # Application du pipeline numérique aux attributs numériques
    ("cat", cat_pipeline, cat_attribs),             # Application du pipeline catégoriel aux attributs catégoriels
])

# Préparation des données : application du transformateur complet aux données
data_prepared = full_pipeline.fit_transform(data)

# Affichage des données originales pour comparaison (facultatif)
print(data_prepared)
```

# Creation du jeu de test et d'entraînement

La division du dataset en deux parties (entraînement et test) est essentielle pour évaluer la capacité du modèle à généraliser. En effet, si le modèle est testé sur les mêmes données que celles utilisées pour l'entraînement, l'évaluation pourrait être biaisée, car le modèle risque de mémoriser les exemples au lieu d'apprendre des motifs généralisables.

---

#### **Répartition choisie :**

| Jeu d'entraînement | Jeu de test |
| ------------------- | ----------- |
| 80%                | 20%         |

---

#### **Échantillonnage stratifié :**
Pour effectuer cette séparation, nous utilisons la technique de **l'échantillonnage stratifié** basée sur la variable **`Attrition`**. Cela garantit que la proportion d'employés ayant quitté l'entreprise est préservée dans les deux jeux de données, offrant ainsi une meilleure représentativité et une évaluation plus fiable des performances du modèle.

```python
from sklearn.model_selection import train_test_split

# 80% pour le jeu d'entraînement et 20% pour le jeu de test
X_train, X_test, y_train, y_test = train_test_split(
    data_prepared, # vos données préparées
    complete_csv["Attrition"], # la variable cible pour la stratification
    test_size=0.2, # la taille du jeu de test
    random_state=42, # le générateur de nombres aléatoires pour la reproductibilité
    stratify=complete_csv["Attrition"] # spécifie la variable cible pour la stratification
)
```

On vérifie que nos données on été séparées en respectant la proportion d'attrition

```python
import matplotlib.pyplot as plt

# Calcul du taux d'attrition pour le jeu d'entraînement
taux_attrition_train = y_train.mean() * 100
# Calcul du taux d'attrition pour le jeu de test
taux_attrition_test = y_test.mean() * 100

# Affichage des résultats
print("Taux d'attrition dans le jeu d'entraînement: {:.2f}%".format(taux_attrition_train))
print("Taux d'attrition dans le jeu de test: {:.2f}%".format(taux_attrition_test))

# Couleurs personnalisées
colors = ["#4CAF50", "#FF9800"]  # Couleurs pour Train et Test

# Création d'un histogramme
plt.figure(figsize=(10, 6))
plt.bar(["No (Train)", "Yes (Train)"], y_train.value_counts(normalize=True) * 100, color=colors[0], alpha=0.8, label="Train", width=0.4)
plt.bar(["No (Test)", "Yes (Test)"], y_test.value_counts(normalize=True) * 100, color=colors[1], alpha=0.8, label="Test", width=0.4)

# Ajout des titres et labels
plt.title("Répartition de l'Attrition entre les jeux d'entraînement et de test", fontsize=16)
plt.xlabel("Attrition", fontsize=14)
plt.ylabel("Pourcentage d'individus (%)", fontsize=14)
plt.legend(loc="upper right", fontsize=12)

# Affichage
plt.tight_layout()
plt.show()
```

#### 1️⃣ **Proportions d'attrition :**
- **Taux d'attrition dans le jeu d'entraînement :** 16.13%.
- **Taux d'attrition dans le jeu de test :** 16.10%.
- Ces proportions sont presque identiques, ce qui confirme que l'échantillonnage stratifié a bien préservé la distribution de la variable cible **`Attrition`** dans les deux jeux.

---

### **Conclusion :**
La répartition des données est fidèle à la distribution initiale. Cela garantit une évaluation fiable des performances des modèles, sans déséquilibres entre les jeux d'entraînement et de test.
