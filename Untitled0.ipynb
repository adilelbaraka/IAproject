Nous avons re√ßu un ensemble de donn√©es du service des ressources humaines sous forme de fichiers CSV. Ces donn√©es, enti√®rement anonymis√©es, garantissent l‚Äôidentification unique de chaque employ√© gr√¢ce √† un identifiant **EmployeeID**, qui est constant √† travers tous les fichiers.

Les fichiers disponibles sont les suivants :

Nous allons commencer par importer ces fichiers afin de les pr√©parer pour l‚Äôanalyse.

```python
from urllib.request import urlretrieve
import pandas as pd
import os

DATASET_URL = "https://raw.githubusercontent.com/adilelbaraka/IAproject/main/dataset/"

DATASET_PATH = "/content/datasets/"
if not os.path.exists(DATASET_PATH):
  os.mkdir(DATASET_PATH)


# Import des fichiers csv (datasets) stock√©s dans un repo github
def import_data(file_name):
  urlretrieve(DATASET_URL+file_name, DATASET_PATH+file_name)
  return pd.read_csv(DATASET_PATH+file_name)

survey = import_data("employee_survey_data.csv")
general = import_data("general_data.csv")
in_time = import_data("in_time.csv")
out_time = import_data("out_time.csv")
manager_survey = import_data("manager_survey_data.csv")
```

## Traitement des donn√©es des fichiers "in_time" et "out_time"

Les fichiers `in_time.csv` et `out_time.csv` ne sont pas directement exploitables, car notre analyse ne porte pas sur les horaires d'arriv√©e et de d√©part des employ√©s.

Cependant, apr√®s examen de ces donn√©es, nous avons identifi√© **deux indicateurs cl√©s** √† extraire :

- **La dur√©e moyenne de travail journali√®re** pour l'ann√©e **2015**.
- **Le nombre total d'absences** sur la m√™me p√©riode.

Ces informations seront consolid√©es dans un **nouveau fichier**, `duration_data.csv`, qui regroupera ces donn√©es essentielles pour l'analyse.

```python
def calculate_duration(in_time, out_time):
    for col in in_time.columns[1:]:
        in_time[col] = pd.to_datetime(in_time[col])
        out_time[col] = pd.to_datetime(out_time[col])

    # Calculer la dur√©e
    duration = in_time.copy()
    for col in in_time.columns[1:]:
        duration[col] = out_time[col] - in_time[col]
        duration[col] = duration[col].dt.total_seconds() / 3600  # Convertir en heures

    return duration


# Calcul de la dur√©e
duration_data = calculate_duration(in_time, out_time)

# Calculer s√©par√©ment les nouvelles colonnes
average_duration = duration_data.iloc[:, 1:].mean(axis=1, skipna=True)
absences_count = duration_data.iloc[:, 1:-1].isna().sum(axis=1)

# Concat√©ner les nouvelles colonnes avec le DataFrame original
duration_data = pd.concat([duration_data, average_duration.rename('Average Duration'),
                           absences_count.rename('Absences Count')], axis=1)

duration_data = duration_data.rename(columns={'Unnamed: 0': "EmployeeID"})

duration_data = duration_data[['EmployeeID','Average Duration', 'Absences Count']]

# Exportation des donn√©es
duration_data.to_csv("duration_data.csv", index=False)
```

# üîç Identification et √©limination des variables non explicatives

#### üìå Objectif : Comprendre la non-explicativit√©
Dans le cadre de notre projet, il est essentiel de d√©terminer quels attributs influencent significativement l'attrition, qui est le param√®tre central de notre analyse. L‚Äôobjectif est donc d‚Äôidentifier et d‚Äô√©liminer les variables non explicatives, c'est-√†-dire celles qui n‚Äôapportent pas de valeur pr√©dictive pour l‚Äôattrition.

#### üß© Crit√®re de non-explicativit√©
Une variable est consid√©r√©e comme **non explicative** si elle pr√©sente une distribution similaire entre les deux groupes d‚Äôattrition ("Oui" et "Non"). Une telle variable n‚Äôa pas de variation significative entre ces groupes et n‚Äôest donc pas un bon indicateur pour pr√©dire l‚Äôattrition.

#### üìä Analyse des variables num√©riques
Pour identifier les variables num√©riques non explicatives, nous utiliserons des **bo√Ætes √† moustaches** (boxplots). Ces visualisations permettent de comparer les distributions des variables entre les employ√©s qui quittent l‚Äôentreprise et ceux qui restent.

- Si les distributions d‚Äôune variable sont similaires entre les deux groupes, elle sera consid√©r√©e comme **non significative** pour l‚Äôanalyse de l‚Äôattrition et pourra √™tre √©limin√©e.
- En revanche, si une variable montre des diff√©rences marqu√©es entre les groupes, elle sera conserv√©e comme attribut explicatif.

Cette approche garantit que seules les variables ayant un impact significatif sur l‚Äôattrition seront prises en compte dans les mod√®les pr√©dictifs.

### üîç Variables cat√©gorielles non explicatives

Pour les variables cat√©gorielles, nous appliquerons une m√©thodologie diff√©rente. Seules les variables montrant des diff√©rences significatives dans leur distribution entre les employ√©s ayant quitt√© l‚Äôentreprise et ceux qui sont rest√©s seront retenues. Ces diff√©rences refl√®tent leur potentiel explicatif en lien avec l‚Äôattrition.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Liste des attributs cat√©goriels
cat_attribs = ["BusinessTravel", "Department", "Gender", "JobRole", "MaritalStatus", "Over18", "EducationField"]

# Configuration pour afficher plusieurs graphiques
fig, axes = plt.subplots(nrows=len(cat_attribs), ncols=1, figsize=(10, 6*len(cat_attribs)))
fig.tight_layout(pad=6.0)

for i, attribut in enumerate(cat_attribs):
    # Calculer les pourcentages
    percentage_df = complete_csv.groupby(['Attrition', attribut]).size().unstack(fill_value=0).apply(lambda x: x / x.sum() * 100, axis=1)

    # Tracer le diagramme en barres sur le subplot correspondant
    ax = axes[i]
    percentage_df.plot(kind='bar', stacked=True, ax=ax)
    ax.set_title(f'Pourcentage de {attribut} par Attrition')
    ax.set_ylabel('Pourcentage (%)')
    ax.set_xlabel('Attrition')

    # Afficher les pourcentages sur les barres
    for n, x in enumerate([*percentage_df.index.values]):
        for (proportion, y_loc) in zip(percentage_df.loc[x], percentage_df.loc[x].cumsum()):
            ax.text(x=n-0.1, y=y_loc - (proportion / 2), s=f'{proportion:.1f}%', color='white', fontsize=10)

    ax.set_xticklabels(['No', 'Yes'])
    ax.legend(title=attribut, bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

### Transformation des donn√©es cat√©gorielles non ordinales pour l'analyse

Les donn√©es cat√©gorielles regroupent des √©l√©ments en diff√©rents groupes distincts. Dans notre ensemble de donn√©es, des colonnes telles que **`Department`** et **`JobRole`** illustrent ce type de donn√©es. Elles sont qualifi√©es de **non ordinales** lorsqu'il n'existe aucun ordre ou hi√©rarchie naturelle entre les cat√©gories. Par exemple, pour la colonne **`Department`**, il n'y a aucune logique √† consid√©rer "Sales" comme sup√©rieur ou inf√©rieur √† "Research & Development".

### Transformation via One-Hot Encoding

Pour rendre ces donn√©es utilisables dans les mod√®les de machine learning, qui n√©cessitent des donn√©es num√©riques, nous appliquons la m√©thode du **One-Hot Encoding**. Cette technique convertit les cat√©gories en vecteurs num√©riques compos√©s de **0** et de **1**, tout en √©vitant d'introduire une hi√©rarchie artificielle.

Voici comment cela fonctionne :
- Chaque cat√©gorie unique d'une colonne est transform√©e en une nouvelle colonne distincte.
- Pour chaque observation, la colonne correspondant √† sa cat√©gorie sera marqu√©e par un **1**, tandis que les autres colonnes seront remplies de **0**.
- Ce processus est appliqu√© √† toutes les colonnes cat√©gorielles concern√©es, assurant ainsi une repr√©sentation claire et exploitable des donn√©es non ordinales.

Cette transformation garantit que les mod√®les de machine learning traitent les donn√©es cat√©gorielles sans biais li√© √† une relation d‚Äôordre inexistante.








```python
from sklearn.preprocessing import OneHotEncoder

cat_encoder = OneHotEncoder()
donnees_cat_1hot = cat_encoder.fit_transform(donnees_cat)
donnees_cat_1hot # array creux par d√©faut
```

Nous appliquerons cette fonction dans notre pipeline

### Cr√©ation de la pipeline

Nous pouvons formater les donn√©es de notre pipeline avec nos donn√©es format√©es:

```python
# Importation des biblioth√®ques n√©cessaires
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np

data = complete_csv.drop("Attrition", axis=1)

# S√©lection des caract√©ristiques num√©riques
donnees_num = data.select_dtypes(include=[np.number])
num_attribs = list(donnees_num)

# Les caract√©ristiques cat√©gorielles ont d√©j√† √©t√© s√©lectionn√©es

# Filtrage des attributs num√©riques, en enlevant ceux sp√©cifi√©s dans to_remove_num
num_attribs = [attr for attr in num_attribs if attr not in to_remove_num]
print(num_attribs)

# Pipeline pour les donn√©es num√©riques
num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),  # Remplissage des valeurs manquantes par la m√©diane
    ('std_scaler', StandardScaler()),               # Standardisation des caract√©ristiques
])

# Pipeline pour les donn√©es cat√©gorielles
cat_pipeline = Pipeline([
    ('onehot_encoder', OneHotEncoder()),            # Encodage One-Hot pour les caract√©ristiques cat√©gorielles
])

# Assemblage des pipelines num√©riques et cat√©gorielles dans un seul transformateur
full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),             # Application du pipeline num√©rique aux attributs num√©riques
    ("cat", cat_pipeline, cat_attribs),             # Application du pipeline cat√©goriel aux attributs cat√©goriels
])

# Pr√©paration des donn√©es : application du transformateur complet aux donn√©es
data_prepared = full_pipeline.fit_transform(data)

# Affichage des donn√©es originales pour comparaison (facultatif)
print(data_prepared)
```

# Creation du jeu de test et d'entra√Ænement

La division du dataset en deux parties (entra√Ænement et test) est essentielle pour √©valuer la capacit√© du mod√®le √† g√©n√©raliser. En effet, si le mod√®le est test√© sur les m√™mes donn√©es que celles utilis√©es pour l'entra√Ænement, l'√©valuation pourrait √™tre biais√©e, car le mod√®le risque de m√©moriser les exemples au lieu d'apprendre des motifs g√©n√©ralisables.

---

#### **R√©partition choisie :**

| Jeu d'entra√Ænement | Jeu de test |
| ------------------- | ----------- |
| 80%                | 20%         |

---

#### **√âchantillonnage stratifi√© :**
Pour effectuer cette s√©paration, nous utilisons la technique de **l'√©chantillonnage stratifi√©** bas√©e sur la variable **`Attrition`**. Cela garantit que la proportion d'employ√©s ayant quitt√© l'entreprise est pr√©serv√©e dans les deux jeux de donn√©es, offrant ainsi une meilleure repr√©sentativit√© et une √©valuation plus fiable des performances du mod√®le.

```python
from sklearn.model_selection import train_test_split

# 80% pour le jeu d'entra√Ænement et 20% pour le jeu de test
X_train, X_test, y_train, y_test = train_test_split(
    data_prepared, # vos donn√©es pr√©par√©es
    complete_csv["Attrition"], # la variable cible pour la stratification
    test_size=0.2, # la taille du jeu de test
    random_state=42, # le g√©n√©rateur de nombres al√©atoires pour la reproductibilit√©
    stratify=complete_csv["Attrition"] # sp√©cifie la variable cible pour la stratification
)
```

On v√©rifie que nos donn√©es on √©t√© s√©par√©es en respectant la proportion d'attrition

```python
import matplotlib.pyplot as plt

# Calcul du taux d'attrition pour le jeu d'entra√Ænement
taux_attrition_train = y_train.mean() * 100
# Calcul du taux d'attrition pour le jeu de test
taux_attrition_test = y_test.mean() * 100

# Affichage des r√©sultats
print("Taux d'attrition dans le jeu d'entra√Ænement: {:.2f}%".format(taux_attrition_train))
print("Taux d'attrition dans le jeu de test: {:.2f}%".format(taux_attrition_test))

# Couleurs personnalis√©es
colors = ["#4CAF50", "#FF9800"]  # Couleurs pour Train et Test

# Cr√©ation d'un histogramme
plt.figure(figsize=(10, 6))
plt.bar(["No (Train)", "Yes (Train)"], y_train.value_counts(normalize=True) * 100, color=colors[0], alpha=0.8, label="Train", width=0.4)
plt.bar(["No (Test)", "Yes (Test)"], y_test.value_counts(normalize=True) * 100, color=colors[1], alpha=0.8, label="Test", width=0.4)

# Ajout des titres et labels
plt.title("R√©partition de l'Attrition entre les jeux d'entra√Ænement et de test", fontsize=16)
plt.xlabel("Attrition", fontsize=14)
plt.ylabel("Pourcentage d'individus (%)", fontsize=14)
plt.legend(loc="upper right", fontsize=12)

# Affichage
plt.tight_layout()
plt.show()
```

- **Taux d'attrition dans le jeu d'entra√Ænement :** 16.13%.
- **Taux d'attrition dans le jeu de test :** 16.10%.
- Ces proportions sont presque identiques, ce qui confirme que l'√©chantillonnage stratifi√© a bien pr√©serv√© la distribution de la variable cible **`Attrition`** dans les deux jeux.
